```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = 'scans_densidad_reliquia/'
pathImagenes = '/home/gustavo/micromegas_5.0.9/imagenes_idms/'
#Los valores reportados por Planck
val_exp = 0.12 #Valor observado
error = 0.001 #Error reportado por Planck 
#Queremos que el valor sea calculado a 3 sigma, o sea, 3 veces el error reportado por Planck 
sigma = 3
val_por = sigma *  error 
val_min = round(val_exp - val_por,3)
val_max = round(val_exp + val_por,3)
```

```python
df1 = pd.read_csv(f'{path}analisis_delta0p01_alpha0p5_Ms5_epsilon1e4.csv')
df2 = pd.read_csv(f'{path}analisis_delta0p01_alpha0p5_Ms5_epsilon1e5.csv')
df3 = pd.read_csv(f'{path}analisis_delta0p01_alpha0p5_Ms500_epsilon1e4.csv')
df4 = pd.read_csv(f'{path}analisis_delta0p01_alpha0p5_Ms500_epsilon1e5.csv')

texto_label = r'$\delta = 0.01 \text{, } \alpha_{D} = 0.5 \text{, } M_{A^\prime} = 3 M_{\chi_{1}}$'
label_1 = r'$M_{\phi} = 5 \text{MeV, } \epsilon = 1\times10^{-4}$'
label_2 = r'$M_{\phi} = 5 \text{MeV, } \epsilon = 1\times10^{-5}$' 
label_3 = r'$M_{\phi} = 500 \text{MeV, } \epsilon = 1\times10^{-4}$' 
label_4 = r'$M_{\phi} = 500 \text{MeV, } \epsilon = 1\times10^{-5}$' 

plt.figure(figsize=(10, 7))

plt.plot(df1['Mchi1']*1e3, df1['Densidad reliquia'], '-', linewidth=2, label=label_1, color='red')
plt.plot(df2['Mchi1']*1e3, df2['Densidad reliquia'], '-', linewidth=2, label=label_2,color='blue')
plt.plot(df3['Mchi1']*1e3, df3['Densidad reliquia'], '--', linewidth=2, label=label_3,color='red')
plt.plot(df4['Mchi1']*1e3, df4['Densidad reliquia'], '--', linewidth=2, label=label_4,color='blue')
plt.fill_between(df1['Mchi1']*1e3 , 
                 val_min, 
                 y2=val_max,  
                 color='magenta', 
                 alpha=0.7)

plt.xscale('log')
plt.yscale('log')

# Ejes y ticks
plt.xticks(fontsize=14)  
plt.yticks(fontsize=14)
plt.tick_params(which='both', direction='in', top=True, right=True)
plt.tick_params(which='major', length=6)
plt.tick_params(which='minor', length=3)

plt.xlim(1, 1000)
plt.ylim(1e-6,1e2)

# Cuadrcula ms sutil
plt.grid(True, linestyle='--', alpha=0.2, which='major')
plt.grid(True, linestyle=':', alpha=0.1, which='minor')

# Etiquetas de ejes
plt.ylabel(r'$\Omega h^2$', size=18, labelpad=10)
plt.xlabel(r'$M_{\chi_1}$ (MeV)', size=18, labelpad=10)

# CUADRO DE TEXTO CON ESPACIO DE PARMETROS (posicin mejorada)
'''
param_text = (
    r'$\alpha_D = 0.5$' + '\n' +
    r'$R = 5$'
)

plt.text(0.95, 0.95, param_text, transform=plt.gca().transAxes,
         fontsize=14, 
         verticalalignment='top',    # Cambiado a top
         horizontalalignment='right', # Cambiado a right
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.95, 
                  edgecolor='black', linewidth=0.8),
         linespacing=1.5)
'''
# Leyenda mejorada
plt.legend(fontsize=14, framealpha=0.95, loc='upper left',
           fancybox=True, shadow=False, frameon=True,
           edgecolor='black', facecolor='white', 
           handletextpad=0.5, handlelength=1.5)

# Marco del grfico
ax = plt.gca()
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_linewidth(0.8)
ax.spines['left'].set_linewidth(0.8)

plt.tight_layout()

# Configuracin global mejorada para publicacin
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman', 'DejaVu Serif'],
    'font.size': 12,
    'axes.labelsize': 16,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
    'legend.fontsize': 14,
    'figure.dpi': 300,
    'savefig.dpi': 300,
})
plt.savefig(f'{pathImagenes}alpha0p5_delta0p01_R3.jpg', dpi=300)
plt.show()
```

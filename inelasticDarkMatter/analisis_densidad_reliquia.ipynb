```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import funcion_l as fl
```

```python
data = {
        'MAp': 3., 
        'mphi': 1,
        'Mchi1': 0.1,
        'angle': 1e-4,
        'gX': 0.1182,
        'epsilon': 0.1,
        'ff': 0.10
    }
```

```python
# Constantes
delta_rel_deseado= 1e-2 #Delta relativo
theta = 5e-4  # Ángulo de mezcla fijo
alphaD = 0.5 #Valor de alpha D
gx = np.sqrt(4*np.pi*alphaD) #Calculo del gx
Ms2 = 5e-2 #Valor de la masa de Higgs
epsilon = 1e-4 #Valor de epsilon
R = 3  #Valor del R

Mx1 = 4e-3
MAp = R * Mx1 #Masa del fotón oscuro
vphi = MAp / (2 * gx) #Valor esperado del escalar
yf = (delta_rel_deseado*Mx1) / (2*vphi)

data2 = data.copy()

data2.update({
            'MAp': MAp,
            'mphi': Ms2,
            'Mchi1': Mx1,
            'angle': theta,
            'gX': gx,
            'epsilon': epsilon,
            'ff': yf})

ob = fl.Likelihood(data2)
dat = ob.get_datos()
dat
```

## Estructura básica de ejecución

```python
start = np.log10(1e-3)  # -3
stop = np.log10(1e0)    # 0
num_points = 10000
```

```python
'''
Al modificar el espacio de parámetros es necesario cambiar
el nombre data3 y el nombre datos3. Junto con los respectivos parámetros.
'''

# Constantes
delta_rel_deseado= 1e-2 #Delta relativo
theta = 5e-4  # Ángulo de mezcla fijo
alphaD = 0.5 #Valor de alpha D
gx = np.sqrt(4*np.pi*alphaD) #Calculo del gx
Ms2_1 = 5e-3 #Valor de la masa de Higgs
Ms2_2 = 5e-1 #Valor de la masa de Higgs
epsilon_1 = 1e-4 #Valor de epsilon
epsilon_2 = 1e-5 #Valor de epsilon
R = 3  #Valor del R
Mx1 = np.logspace(-3,0,2) #Valores de masa
datos_1 = [] #Cambiar datos3 
datos_2 = [] 
datos_3 = []
datos_4 = []
data = {'MAp': 3., 
        'mphi': 1,
        'Mchi1': 0.1,
        'angle': theta,
        'gX': gx,
        'epsilon': 0.1,
        'ff': 0.10}

data_1 = data.copy()
data_2 = data.copy()
data_3 = data.copy()
data_4 = data.copy()

data_1['mphi'] = Ms2_1
data_1['epsilon'] = epsilon_1
data_2['mphi'] = Ms2_1
data_2['epsilon'] = epsilon_2
data_3['mphi'] = Ms2_2
data_3['epsilon'] = epsilon_1
data_4['mphi'] = Ms2_2
data_4['epsilon'] = epsilon_2



for x in np.logspace(start, stop, num=num_points):
    MAp = R * x
    vphi = MAp / (2 * gx)
    yf = (delta_rel_deseado*x) / (2*vphi)
    
    # Para data_1
    current_data_1 = data_1.copy()
    current_data_1['MAp'] = MAp
    current_data_1['Mchi1'] = x
    current_data_1['ff'] = yf
    ob_1 = fl.Likelihood(current_data_1)
    dat_1 = ob_1.get_datos()
    datos_1.append(dat_1)
    
    # Para data_2
    current_data_2 = data_2.copy()
    current_data_2['MAp'] = MAp
    current_data_2['Mchi1'] = x
    current_data_2['ff'] = yf
    ob_2 = fl.Likelihood(current_data_2)
    dat_2 = ob_2.get_datos()
    datos_2.append(dat_2)

    # Para data_3
    current_data_3 = data_3.copy()
    current_data_3['MAp'] = MAp
    current_data_3['Mchi1'] = x
    current_data_3['ff'] = yf
    ob_3 = fl.Likelihood(current_data_3)
    dat_3 = ob_3.get_datos()
    datos_3.append(dat_3)

    # Para data_4
    current_data_4 = data_4.copy()
    current_data_4['MAp'] = MAp
    current_data_4['Mchi1'] = x
    current_data_4['ff'] = yf
    ob_4 = fl.Likelihood(current_data_4)
    dat_4 = ob_4.get_datos()
    datos_4.append(dat_4)
    
df1 = pd.DataFrame(datos_1)
df1.to_csv(f'{path}analisis_delta0p01_alpha0p5_Ms5_epsilon1e4.csv')
df2 = pd.DataFrame(datos_2)
df2.to_csv(f'{path}analisis_delta0p01_alpha0p5_Ms5_epsilon1e5.csv')
df3 = pd.DataFrame(datos_3)
df3.to_csv(f'{path}analisis_delta0p01_alpha0p5_Ms500_epsilon1e4.csv')
df4 = pd.DataFrame(datos_4)
df4.to_csv(f'{path}analisis_delta0p01_alpha0p5_Ms500_epsilon1e5.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def densidad(x,delta):
    return np.exp(-(delta+1)*x)

def densidad_chi1(x): 
    return np.exp(-x)
    
x = np.arange(1,100,1)
t1 = 10 / x 
t2 = 100 / x
t3 = 1000 / x 
y = densidad(x, 0.1)
y2 = densidad_chi1(x)

# Constantes fundamentales
ħ = 1.0545718e-34  # J·s (constante de Planck reducida)
k_B = 1.380649e-23 # J/K (constante de Boltzmann)

def temperatura_a_tiempo(temperatura_K):
    """
    Convierte temperatura en Kelvin a tiempo característico en segundos
    """
    energia_J = k_B * temperatura_K  # Energía térmica en Joules
    tiempo_s = ħ / energia_J        # Tiempo característico en segundos
    return tiempo_s
```

```python
plt.figure()
plt.plot(temperatura_a_tiempo(t1),y2, '--', color='black')
plt.plot(temperatura_a_tiempo(t1),y,'-',color='red')
#plt.plot(temperatura_a_tiempo(t2),y,'--',color='red')
#plt.plot(temperatura_a_tiempo(t3),y,'--',color='blue')
plt.xscale('log')
plt.yscale('log')
#plt.gca().invert_xaxis()  # Eje x invertido
plt.xlabel('Tiempo (s)')
plt.ylabel(r'Densidad numerica $n_{2}$')
plt.show()
```

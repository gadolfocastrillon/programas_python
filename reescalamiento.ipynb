```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Formato de notación scientifica
plt.ticklabel_format(axis='both', style='sci', scilimits=(0,0))
# o específicamente para cada eje
plt.ticklabel_format(axis='x', style='sci', scilimits=(0,0))
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
```

```python
na64 = 'limitesidms/NA64/'
#https://inspirehep.net/files/83ac756d095b4e927d0afa0715ff8a07 
limite_na64_delta0p1_alpha0p5_R10 = pd.read_csv(f'{na64}NA64_delta0p1_alpha0p5_R10.csv',sep=';',decimal=',', names=['Mx1','y'])
limite_na64_delta0p4_alpha0p1_R3 = pd.read_csv(f'{na64}NA64_delta0p4_alpha0p1_R3.csv',sep=';',decimal=',', names=['Mx1','y'])
limite_na64_delta0p4_alpha0p5_R3 = pd.read_csv(f'{na64}NA64_delta0p4_alpha0p5_R3.csv',sep=';',decimal=',', names=['Mx1','y'])
limite_na64_delta0p1_alpha0p1_R3 = pd.read_csv(f'{na64}NA64_delta0p1_alpha0p1_R3.csv',sep=';',decimal=',', names=['Mx1','y'])
limite_na64_delta0p1_alpha0p5_R10.y = limite_na64_delta0p1_alpha0p5_R10.y
```

```python
factor1 = 0.1 * (1/3)**4
factor2 = 0.5 * (1/3)**4
plt.figure()
limite_na64_delta0p4_alpha0p5_R3 = limite_na64_delta0p4_alpha0p5_R3.sort_values('Mx1')
plt.plot(limite_na64_delta0p4_alpha0p5_R3.Mx1*1e3, limite_na64_delta0p4_alpha0p5_R3.y / factor2,'.',color='blue',label=r'$\alpha = 0.5$')
plt.plot(limite_na64_delta0p4_alpha0p1_R3.Mx1*1e3, limite_na64_delta0p4_alpha0p1_R3.y / factor1,'--',color='red', label=r'$\alpha = 0.1$')
#plt.plot(limite_na64_delta0p1_alpha0p5_R10.Mx1*1e3, limite_na64_delta0p1_alpha0p5_R10.y / factor,'--',color='black')
#-------------------------------------------------------------------------------------------------
#                                      Elementos gráficos mejorados
#-------------------------------------------------------------------------------------------------
plt.xscale('log')
plt.yscale('log')

# Ejes y ticks
plt.xticks(fontsize=14)  
plt.yticks(fontsize=14)
plt.tick_params(which='both', direction='in', top=True, right=True)
plt.tick_params(which='major', length=6)
plt.tick_params(which='minor', length=3)

plt.xlim(1, 1000)
plt.ylim(1e-14,1e-6)

# Cuadrícula más sutil
plt.grid(True, linestyle='--', alpha=0.2, which='major')
plt.grid(True, linestyle=':', alpha=0.1, which='minor')

# Etiquetas de ejes
plt.ylabel(r'$\epsilon^2$', size=18, labelpad=10)
plt.xlabel(r'$M_{\chi_1}$ (MeV)', size=18, labelpad=10)

# CUADRO DE TEXTO CON ESPACIO DE PARÁMETROS (posición mejorada)
param_text = (
    r'$M_{A^\prime} = 3 M_{\chi_{1}}$'
)

plt.text(0.05, 0.95, param_text, transform=plt.gca().transAxes,
         fontsize=14, 
         verticalalignment='top',    # Mantener top
         horizontalalignment='left', # Cambiado a left
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.95, 
                  edgecolor='black', linewidth=0.8),
         linespacing=1.5)

# Marco del gráfico
ax = plt.gca()
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_linewidth(0.8)
ax.spines['left'].set_linewidth(0.8)

plt.tight_layout()

plt.legend(fontsize=14, framealpha=0.95, loc='lower right',
           fancybox=True, shadow=False, frameon=True,
           edgecolor='black', facecolor='white', 
           handletextpad=0.5, handlelength=1.5)

# Configuración global mejorada para publicación
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman', 'DejaVu Serif'],
    'font.size': 12,
    'axes.labelsize': 16,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
    'legend.fontsize': 14,
    'figure.dpi': 300,
    'savefig.dpi': 300,
})

#plt.savefig('imagenes_idms/Delta0p05_alpha0p5_R5_separado.jpg', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

Aquí se puede observar como, para $\alpha_D = 0.5 \text{ y } 0.1$ dan exactamente el mismo gráfico. Esto hace que sea posible utilizar los resultados de $\alpha = 0.1$ para los datos que tenemos en $\alpha = 0.5$.   

```python
factor = 0.1 * (1/3)**4
plt.figure()
plt.plot(limite_na64_delta0p1_alpha0p1_R3.Mx1*1e3, limite_na64_delta0p1_alpha0p1_R3.y / factor,'-',color='blue',label=r'$\delta = 0.1$')
plt.plot(limite_na64_delta0p4_alpha0p1_R3.Mx1*1e3, limite_na64_delta0p4_alpha0p1_R3.y / factor,'--',color='red', label=r'$\delta = 0.4$')
#plt.plot(limite_na64_delta0p1_alpha0p5_R10.Mx1*1e3, limite_na64_delta0p1_alpha0p5_R10.y / factor,'--',color='black')
#-------------------------------------------------------------------------------------------------
#                                      Elementos gráficos mejorados
#-------------------------------------------------------------------------------------------------
plt.xscale('log')
plt.yscale('log')

# Ejes y ticks
plt.xticks(fontsize=14)  
plt.yticks(fontsize=14)
plt.tick_params(which='both', direction='in', top=True, right=True)
plt.tick_params(which='major', length=6)
plt.tick_params(which='minor', length=3)

plt.xlim(1, 1000)
plt.ylim(1e-14,1e-6)

# Cuadrícula más sutil
plt.grid(True, linestyle='--', alpha=0.2, which='major')
plt.grid(True, linestyle=':', alpha=0.1, which='minor')

# Etiquetas de ejes
plt.ylabel(r'$\epsilon^2$', size=18, labelpad=10)
plt.xlabel(r'$M_{\chi_1}$ (MeV)', size=18, labelpad=10)

# CUADRO DE TEXTO CON ESPACIO DE PARÁMETROS (posición mejorada)
param_text = (
    r'$\alpha_D = 0.1$' + '\n' +
    r'$M_{A^\prime} = 3 M_{\chi_{1}}$'
)

plt.text(0.05, 0.95, param_text, transform=plt.gca().transAxes,
         fontsize=14, 
         verticalalignment='top',    # Mantener top
         horizontalalignment='left', # Cambiado a left
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.95, 
                  edgecolor='black', linewidth=0.8),
         linespacing=1.5)

# Marco del gráfico
ax = plt.gca()
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_linewidth(0.8)
ax.spines['left'].set_linewidth(0.8)

plt.tight_layout()

plt.legend(fontsize=14, framealpha=0.95, loc='lower right',
           fancybox=True, shadow=False, frameon=True,
           edgecolor='black', facecolor='white', 
           handletextpad=0.5, handlelength=1.5)

# Configuración global mejorada para publicación
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman', 'DejaVu Serif'],
    'font.size': 12,
    'axes.labelsize': 16,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
    'legend.fontsize': 14,
    'figure.dpi': 300,
    'savefig.dpi': 300,
})

#plt.savefig('imagenes_idms/Delta0p05_alpha0p5_R5_separado.jpg', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```

Basicamente lo que estamos observando, es que, para bajos valores de $\epsilon$ menores a $1\times10^{-6}$ no tenemos ningún cambio entre $\delta = 0.1$ y $\delta = 0.4$.  Es posible por lo tanto, para $\alpha = 0.1$ y $R = 3$  establecer limites para $\delta = 0.1 \text{, } 0.2 \text{ y } 0.4 $. Y debido a la naturaleza del experimento, es posible plantear que, para $R = 10$ es posible y $\delta = 0.1$ es posible que este limite sea el mismo para los otros dos deltas buscados (en relación a lo visto anteriormente). 

```python
factor = 0.1 * (1/3)**4
factor2 = 0.5 * (1/10)**4 
plt.figure()
plt.plot(limite_na64_delta0p1_alpha0p1_R3.Mx1*1e3, limite_na64_delta0p1_alpha0p1_R3.y / factor * (0.1*(1/3)**2) ,'-',color='blue',label=r'$R = 3$')
plt.plot(limite_na64_delta0p1_alpha0p5_R10.Mx1*1e3, limite_na64_delta0p1_alpha0p5_R10.y / factor2 * (0.5*(1/10)**2) * (0.1*(1/3)**2),'--',color='black'label=r'$R = 10$')
#-------------------------------------------------------------------------------------------------
#                                      Elementos gráficos mejorados
#-------------------------------------------------------------------------------------------------
plt.xscale('log')
plt.yscale('log')

# Ejes y ticks
plt.xticks(fontsize=14)  
plt.yticks(fontsize=14)
plt.tick_params(which='both', direction='in', top=True, right=True)
plt.tick_params(which='major', length=6)
plt.tick_params(which='minor', length=3)

plt.xlim(1, 1000)
plt.ylim(1e-14,1e-6)

# Cuadrícula más sutil
plt.grid(True, linestyle='--', alpha=0.2, which='major')
plt.grid(True, linestyle=':', alpha=0.1, which='minor')

# Etiquetas de ejes
plt.ylabel(r'$\epsilon^2$', size=18, labelpad=10)
plt.xlabel(r'$M_{\chi_1}$ (MeV)', size=18, labelpad=10)

# CUADRO DE TEXTO CON ESPACIO DE PARÁMETROS (posición mejorada)
param_text = (
    r'$\alpha_D = 0.1$' + '\n' +
    r'$M_{A^\prime} = 3 M_{\chi_{1}}$'
)

plt.text(0.05, 0.95, param_text, transform=plt.gca().transAxes,
         fontsize=14, 
         verticalalignment='top',    # Mantener top
         horizontalalignment='left', # Cambiado a left
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.95, 
                  edgecolor='black', linewidth=0.8),
         linespacing=1.5)

# Marco del gráfico
ax = plt.gca()
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_linewidth(0.8)
ax.spines['left'].set_linewidth(0.8)

plt.tight_layout()

plt.legend(fontsize=14, framealpha=0.95, loc='lower right',
           fancybox=True, shadow=False, frameon=True,
           edgecolor='black', facecolor='white', 
           handletextpad=0.5, handlelength=1.5)

# Configuración global mejorada para publicación
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman', 'DejaVu Serif'],
    'font.size': 12,
    'axes.labelsize': 16,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
    'legend.fontsize': 14,
    'figure.dpi': 300,
    'savefig.dpi': 300,
})

#plt.savefig('imagenes_idms/Delta0p05_alpha0p5_R5_separado.jpg', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
